```python
# package imports go here
import pandas as pd
import numpy as np
import fastparquet as fp
import os
import sys
import pickle

sys.path.insert(1, '../pkgs')
import ml_functions as mlfuncs
```

```python
# Path to results
result_path     = "../../data/"
report_file     = 'feature_report.pkl'
```

```python
diabetes_features = [
    'GENHLTH', 'PHYSHLTH', 'MENTHLTH', 'PRIMINSR', 'PERSDOC3', 'CHECKUP1', 'EXERANY2', 
    'SLEPTIM1', 'CVDCRHD4', 'CVDSTRK3', 'ADDEPEV3', 'CHCKDNY2', 'DIABETE4', 'MARITAL', 
    'EDUCA', 'RENTHOM1', 'EMPLOY1', 'INCOME3', 'WEIGHT2', 'DEAF', 'BLIND', 'DIFFWALK', 
    'FLUSHOT7', 'COVIDPOS', 'COVIDSMP', 'PREDIAB2', 'DIABTYPE', 'CHKHEMO3', 'EYEEXAM1', 
    'DIABEYE1', 'FEETSORE', 'TOLDCFS', 'HAVECFS', 'HPVADVC4', 'SHINGLE2', 'COVIDVA1', 
    'COVIDNU1', 'COPDSMOK', 'CIMEMLOS', 'CDDISCUS', 'MSCODE', '_IMPRACE', '_RFHLTH', 
    '_HLTHPLN', '_TOTINDA', '_MICHD', '_PRACE2', '_RACE1', '_RACEGR4', '_SEX', 
    '_AGEG5YR', '_AGE80', 'WTKG3', '_BMI5', '_BMI5CAT', '_EDUCAG', '_INCOMG1', 
    '_SMOKER3', '_RFSMOK3', '_CURECI2', '_SMOKGRP', '_DRNKWK2', 'BPHIGH6', 'BPMEDS', 
    'CHOLCHK3', 'TOLDHI3', 'CHOLMED3', 'PREDIAB1', 'TOLDHEPC', 'HAVEHEPB']
```

```python
all_features = {}
diabetes_report = []
#for year in range(1990, 2023):
for year in range(2020, 2023):
    if (year != 2009):
        parquet_file = mlfuncs.brfss_parquet_file( result_path, year )
        print(f"Reading File: {parquet_file}")
        df = pd.read_parquet(mlfuncs.brfss_parquet_file( result_path, year ), engine="fastparquet")
        columns = df.columns
        for feature in df.columns:
            if feature in all_features:
                all_features[feature].append(year)
            else:
                all_features[feature] = []
                all_features[feature].append(year)
        
        diabetes_features_year = []
        for feature in diabetes_features:
            if feature in columns:
                diabetes_features_year.append('yes')
            else:
                diabetes_features_year.append('no')
        diabetes_report.append(diabetes_features_year)

report_df = pd.DataFrame(diabetes_report, columns=diabetes_features)
            
all_features['report'] = report_df
```

```python
# Save list/dict/dataframe structure to file
with open(result_path+report_file, 'wb') as file:
    pickle.dump(all_features, file)
```

```python
# Load list/dict/dataframe structure from  file
with open(result_path+report_file, 'rb') as file:
    report_dict = pickle.load(file)
```
